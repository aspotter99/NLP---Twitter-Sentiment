```python
import pandas as pd

df = pd.read_csv('../../data/expanded_dataframe.csv', index_col = 0)
```

```python
import sqlite3 as sl

con = sl.connect('tweets')

df = pd.read_sql('''
    SELECT *
    FROM tweets
    ''', con)
```

```python
df
```

```python
apple = pd.read_sql('''
                    SELECT *
                    FROM tweets
                    WHERE apple_product = '1'
                    ORDER BY pos_rev desc
                    ''',con)
apple
```

```python
sxsw = pd.read_sql('''
                   SELECT *
                   FROM tweets
                   WHERE tweet_text like '%sxsw%'
                   ORDER BY pos_rev desc''',con)
sxsw
```

```python
google = pd.read_sql('''
                    SELECT *
                    FROM tweets
                    WHERE apple_product = '0'
                    ''',con)
google
```

```python
data = pd.read_sql("""
                    SELECT tweet_text, pos_rev, neg_rev, apple_product
                    FROM tweets
                    """,con)
data
```
